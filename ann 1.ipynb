```python
import numpy as np

def sigmoid(x):
    return 1/(1+ np.exp(-x))
```

```python
def relu(x):
    return np.maximum(0,x)
```

```python
def tanh(x):
    return np.tanh(x)
```

```python
def softmax(x):
    exp_x = np.exp(x -np.max(x, axis = -1 ,keepdims =True))
    return exp_x/np.sum(exp_x, axis = -1 ,keepdims =True)               
```

```python
x = np.array([-2, -1, 0, 1, 2])

print("Sigmoid:", sigmoid(x))
print("ReLU:", relu(x))
print("Tanh:", tanh(x))
print("Softmax:", softmax(x))
```

```python
import matplotlib.pyplot as plt
```

```python
Sigmoid = sigmoid(x)
ReLU =relu(x)
Tanh = tanh(x)
Softmax= softmax(np.vstack([x,x+2]))
```

```python
plt.subplot(2,2,1)
plt.plot(x ,Sigmoid,label =Sigmoid)
plt.title('SIGMOID ACTIVATION FUNCTION')
plt.legend()
```

```python
plt.subplot(2,2,1)
plt.plot(x ,ReLU ,label =ReLU)
plt.title('ReLU ACTIVATION FUNCTION')
plt.legend()
```

```python
plt.subplot(2,2,1)
plt.plot(x ,Tanh ,label =Tanh)
plt.title('Tanh ACTIVATION FUNCTION')
plt.legend()
```

```python
plt.subplot(2,2,1)
plt.plot(x ,Softmax[0] ,label ='Softmax(x)')
plt.plot(x ,Softmax[1] ,label ='Softmax(x+2)')
plt.title('Softmax ACTIVATION FUNCTION')
plt.legend()
```
